```python
def solution(participant, completion):
    d = {}
    for x in participant:
        d[x] = d.get(x, 0) + 1   # x가 있으면 원래 있는 갯수 +1, 아니면 0
    for x in completion:
        d[x] -= 1   # participant에 있는 갯수에서 완주한 갯수를 빼주기 ---> 0 이면 완주했다는 뜻
    dnf = [k for k, v in d.items() if v > 0]
    answer = dnf[0]
    return answer
```
